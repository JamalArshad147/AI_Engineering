```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./datasets/Position_Salaries.csv")
df.columns
```

```python
X = df[["Level"]].values
y = df["Salary"].values
```

```python
y = y.reshape(len(y), 1)
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
sc_y = StandardScaler()

X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
from sklearn.svm import SVR

model = SVR(kernel='rbf')
model.fit(X, y)
```

```python
model.predict([[10.5]])
```

```python
sc_y.inverse_transform(model.predict(sc_X.transform([[10.5]])).reshape(-1, 1))
```

```python
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color="g")
plt.plot(sc_X.inverse_transform(X), sc_y.inverse_transform(model.predict(X).reshape(-1, 1)))
plt.title("Truth or Buff(Linear Regression)")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()

```
