```python
import google.generativeai as genai

for model in genai.list_models():
    print(model.name)
```

```python
# Generating text from text prompts

model = genai.GenerativeModel('gemini-2.5-flash')
response = model.generate_content("Explain Agentic AI")
```

```python
print(response.text)
```

```python
model_info = genai.get_model('gemini-2.5-flash')
print(model_info)
```

```python
# Streamming in Gemini

prompt = "Write 1000 words on AI"
response = model.generate_content(prompt, stream=True)
for chunk in response:
    print(chunk.text, end="")
```

```python
import time 

model = genai.GenerativeModel('gemini-2.5-pro')
chat = model.start_chat(history=[])

while True:
    prompt = input('User: ')
    if prompt.lower() not in ['bye', 'exit', 'stop']:
        response = chat.send_message(prompt)
        print(f'{chat.history[-1].role.capitalize()} : {chat.history[-1].parts[0].text}')
        print('\n' + '*'*100 + '\n')
    else:
        print('Saving...')
        time.sleep(4)
        print("Bye bye!")
        break
```
